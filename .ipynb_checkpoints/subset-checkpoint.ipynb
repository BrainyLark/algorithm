```python
import numpy as np
```

```python
class Set:
    
    def __init__(self, initset):
        self.container = np.zeros((2, len(initset)), dtype='int')
        self.container[0, ] = [i for i in range(len(initset))]
        self.container[1, ] = [ord(ch) for ch in initset]
        self.collector = set()
        
    def track(self, keys):
        if len(keys) == 0:
            return
        
        for i in range(len(keys)):
            self.collector.add(tuple(keys))
            self.track(np.append(keys[:i], keys[i+1:]))
        
    def print_all(self):
        temp = []
        for elem in self.collector:
            temp.append(list(elem))
        print("Number of subsets: ", len(temp))
        for unit in range(len(temp)):
            print([chr(self.container[1, elem]) for elem in temp[unit]])
```

```python
s = Set(['a', 'r', 's', 'e', 'n', 'a', 'l'])
s.track(s.container[0, ])
s.print_all()
```
